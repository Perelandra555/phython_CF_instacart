#### 06 
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# calculate the mean  of the prices of products purchased by a user 
#new column ['mean_price']
# group by user_id
ords_prods_merge['mean_prices'] = ords_prods_merge.groupby('user_id')['prices'].transform('mean')
```

```python
ords_prods_merge.head()
```

```python
# create the spender type categories due to the mean prices

ords_prods_merge.loc[ords_prods_merge['mean_prices'] >= 10, 'spender_type'] = 'High spender'

ords_prods_merge.loc[ords_prods_merge['mean_prices'] < 10, 'spender_type'] = 'Low spender'
```

```python
ords_prods_merge[['user_id', 'mean_prices', 'spender_type']].head(60)
```

#### 07  Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.

print(df.median(axis="columns"))

```python
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform('median')
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'customer_frequency'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] <= 20) & (ords_prods_merge['median_days_since_prior_order'] > 10), 'customer_frequency'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'customer_frequency'] = 'Frequent customer'
```
