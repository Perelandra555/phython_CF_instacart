#### 04 The Instacart CFO isn’t interested in customers who don’t generate much revenue for the app. Create an exclusion flag for low-activity customers (customers with less than 5 orders) and exclude them from the data. Make sure you export this sample.

```python
# finding the maximum number of orders for each user
df['max_order_number'] = df.groupby('user_id')['order_number'].transform('max')
```

```python
df.head()
```

```python
# create a function called customer_activity
def customer_activity(row):

  if row['max_order_number'] < 5:
    return 'low-activity customer'
  elif row['max_order_number'] >= 5:
    return 'high-activity customer'
  else: return 'Not enough data'
```

```python
#create a new column for costumer activity
df.loc[:, 'customer_activity'] = df.apply(customer_activity, axis=1)
```
